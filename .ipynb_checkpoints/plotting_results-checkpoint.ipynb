# Plotting the results of the Random Forest Classification

After the successful classification with the trained Random Forest Classifier, we want to display all our results uniformly. The best way to do this is to use Matplotlib. We also plot the distribution of classes in a nice basic pie chart for better visualization. First, we need to import all need packages, as usual.

```python
from osgeo import gdal
import fiona
import numpy as np
import pandas as pd
import geopandas as gpd

import rasterio as rio

from nansat import Nansat
from nansat import *

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

Then, we open all our prediction result, that we saved as Tiff-files (raster format) after the prediction with the classifier. 

```python
feb_pred = Nansat('/container_dir1/results/RFC_S1_pred_feb.tif')
march_pred = Nansat('/container_dir1/results/RFC_S1_pred_march.tif')
april_pred = Nansat('/container_dir1/results/RFC_S1_pred_april.tif')
may_pred = Nansat('/container_dir1/results/RFC_S1_may.tif')
```

```python
june_pred = Nansat('/container_dir1/results/RFC_S1_pred_june.tif')
july_pred_early = Nansat('/container_dir1/results/RFC_S1_july_early.tif')
july_pred_late = Nansat('/container_dir1/results/RFC_S1_pred_july2.tif')
aug_pred = Nansat('/container_dir1/results/RFC_S1_aug.tif')
sep_pred = Nansat('/container_dir1/results/RFC_S1_pred_sep.tif')
nov_pred = Nansat('/container_dir1/results/RFC_S1_pred_nov.tif')
dec_pred = Nansat('/container_dir1/results/RFC_S1_pred_dec.tif')
```

```python
stacked_class = Nansat('/container_dir1/results/RFC_s1s1vis.tif')
S2_pred = Nansat('/container_dir1/results/RFC_S2_pred.tif')
vis_pred = Nansat('/container_dir1/results/RFC_VIs_pred.tif')
```

We store them all in a list, because plotting all those nice predictions we have one after the other would make a lot of code.

```python
pred = (feb_pred[1], march_pred[1], april_pred[1], may_pred[1], june_pred[1], july_pred_early[1], july_pred_late[1],
        aug_pred[1], sep_pred[1], nov_pred[1], dec_pred[1], S2_pred[1], vis_pred[1], stacked_class[1])
```

To iterate over all the Tiff-files in a loop, we have to turn the list into ArtyomÂ´s favourite thing: an array. Hurray!
To check if the array is the same as the list, we plot the February image. That should be the same picture.

```python
pred_arr = np.array(pred)
plt.imshow(feb_pred[1])
```

```python
plt.imshow(pred_arr[0])
```

We should know what we are plotting in the end, so we create a list with labels for the classes (Not Saltmarsh and Saltmarsh) and for the pictures we will plot. 
In the loop we calculate the class distribution for each prediction date (np.count_nonzero - with the respective class ID) and generate pie charts. Since we use the colourmap 'Viridis' for the prediction picture later, we use the colours of this colormap for the pie chart as well.

```python
# Label the classes (labels) and the months (names)
labels = ['Not saltmarsh', 'Saltmarsh']
names = ['February', 'March', 'April', 'May', 'June', 'July 07.07.', 'July 22.07.', 'August', 'September', 'November', 'December', 'S2 July', 'VIs July', 'Stack July']

# Calculate the class distribution for each prediction date and generate a nice basic b pie chart 
for i in range(pred_arr.shape[0]):
    raster_pred = pred_arr[i]
    
    #count the classes
    Y = np.count_nonzero(raster_pred == 1)
    X = np.count_nonzero(raster_pred == 6)
    classes = [Y, X]
    
    # get colors of classification colormap to plot the pie charts
    colormap = plt.cm.viridis
    numberOfSlices = len(classes)
    sliceColors = colormap(np.linspace(0., 1., numberOfSlices))
    
    #plot the distribution
    fig = plt.figure(figsize=(10,10))  
    plt.pie(classes, 
            labels=labels, 
            autopct='%1.1f%%', 
            colors=sliceColors,
            textprops=dict({'fontsize':13}))
    plt.title("Random Forest class distribution "+names[i]+" 2021\n\n"+
              "Saltmarsh: "+str(X)+"\nNot saltmarsh: "+str(Y), fontsize=16)
    
    #Save the plots
    plt.savefig('pie_'+names[i]+'_2021.png')
    
    #Close the plots, so they donÂ´t overwrite each other
    plt.close()
```

Now it is time to plot the classified pictures. 

```python
# Plot classifications, all of them 
for i in range(pred_arr.shape[0]):
    fig = plt.figure(figsize=(50,50))
    plt.axis("off")
    plt.imshow(pred_arr[i])
    plt.title("Binary Random Forest Classification\n"
              +names[i]+" 2021", fontsize=55)
    plt.savefig("RFC_bin_2021_"+names[i]+".png")
    plt.close()
```

To display how a pie chart would look like, one is plotted outside the loop for demonstration.

Find all the figures in the folder /figures.

```python
Y = np.count_nonzero(stacked_class[1] == 1)
X = np.count_nonzero(stacked_class[1] == 6)
classes = [Y, X]
fig = plt.figure(figsize=(10,10))  
plt.pie(classes, 
        labels=labels, 
        autopct='%1.1f%%', 
        colors=sliceColors,
        textprops=dict({'fontsize':13}))
plt.title("Random Forest class distribution Stack 2021\n\n"+
        "Saltmarsh: "+str(X)+"\nNot saltmarsh: "+str(Y), fontsize=16)

plt.savefig('pie_stack_2021.png')
```
